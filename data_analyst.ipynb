## **Oпис моделі та її ефективності**

Використана модель - Random Forest Classifier:

Random Forest — це ансамблева модель, яка використовує множину дерев рішень для підвищення стабільності та точності передбачень. Вона зменшує ризик перенавчання та є особливо ефективною для великих і складних датасетів.

Основні параметри моделі:
 - n_estimators: 100 (кількість дерев в лісі)
 - max_depth: необмежена (дерева будуються до повного розщеплення або до мінімального числа зразків у вузлі)
 - random_state: 42 (для відтворюваності результатів)

***Навчання моделі***

Перед навчанням моделі, дані були попередньо оброблені:
* Обробка пропущених значень: Заповнення пропущених значень за допомогою SimpleImputer (стратегія 'mean').
* Категоріальні змінні: Виконано One-Hot Encoding для категоріальної змінної 'sample_type'.
* Розділення даних: Дані були розділені на тренувальну (dev), валідаційну (valid) і тестову (test) вибірки.

***Метрики моделі:***
 - Accuracy: 0.9539 (точність)
 - F1-score: 0.9547 (гармонійне середнє між точністю та повнотою)
 - AUC-ROC: 0.9848 (площа під кривою ROC, що показує здатність моделі розрізняти між класами)


###**Проблеми, які вирішувалися**

1. **Дисбаланс класів:**

*Проблема:* Клас 'good' значно переважав клас 'bad', що могло призвести до упередженості моделі в бік більш представленого класу.

*Рішення:* Використання метрик, таких як F1-score та AUC-ROC, які краще відображають ефективність моделі на дисбалансованих даних. Крім того, в майбутньому можна застосувати методи балансування даних, такі як oversampling або undersampling.


2. **Пропущені значення:**

*Проблема:* Пропущені значення можуть призвести до помилок у моделі.

*Рішення:* Видалення або заповнення пропущених значень за допомогою середнього значення (SimpleImputer).


3. **Переобучення моделі:**

*Проблема:* Перенавчання може виникати, коли модель запам'ятовує тренувальні дані замість того, щоб навчатися загальним патернам.

*Рішення:* Використання Random Forest, який є менш схильним до перенавчання через свій ансамблевий характер.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, f1_score, roc_curve, roc_auc_score, auc, confusion_matrix
import seaborn as sns
from imblearn.over_sampling import SMOTE
```

```python
!pip install mplcyberpunk
import mplcyberpunk

plt.style.use("cyberpunk")
```

```python
data = pd.read_csv('dataset.csv')
```

```python
# Removing rows with missing values in the 'target' column
data = data.dropna(subset=['target'])

# Analysis and visualization of descriptive statistics
print("Descriptive statistics of the target variable:")
print(data['target'].describe())
```

```python
plt.figure(figsize=(6, 4))
plt.hist(data['target'], bins=20, edgecolor='k')
plt.xlabel('Target')
plt.ylabel('Frequency')
plt.title('Distribution of Target Variable')
plt.show()
```

```python
# Converting the target variable to numeric format
y = data['target'].replace({'good': 1, 'bad': 0})

# Removing columns that are not features
X = data.drop(['target', 'sample_type'], axis=1)

# Handling missing values
imputer = SimpleImputer(strategy='mean')
X_imputed = imputer.fit_transform(X)

# Encoding categorical variables using one-hot encoding
encoder = OneHotEncoder(sparse_output=False, drop='first')
sample_type_encoded = encoder.fit_transform(data[['sample_type']])
X_encoded = np.concatenate([X_imputed, sample_type_encoded], axis=1)

# Standardizing the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_encoded)

# Balancing classes using SMOTE
smote = SMOTE(random_state=42)
X_resampled, y_resampled = smote.fit_resample(X_scaled, y)

# Splitting the data into dev, test, and valid sets
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.2, random_state=42, stratify=y_resampled)
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.25, random_state=42, stratify=y_train)

# Choosing and training the Random Forest model
rf_model = RandomForestClassifier(random_state=42)
rf_model.fit(X_train, y_train)

# Evaluating the model on the test set
y_pred = rf_model.predict(X_test)

accuracy = accuracy_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)
y_test_binary = y_test

# Getting probabilities for the positive class
y_proba = rf_model.predict_proba(X_test)[:, 1]

# Calculating AUC-ROC
auc = roc_auc_score(y_test_binary, y_proba)

print('Random Forest model metrics:')
print('Accuracy:', accuracy)
print('F1-score:', f1)
print('AUC-ROC:', auc)
```

## Крива ROC

```python
fpr, tpr, thresholds = roc_curve(y_test_binary, y_proba)
roc_auc = auc(fpr, tpr)

plt.figure(figsize=(6, 4))
plt.plot(fpr, tpr, color='blue', lw=2, label='ROC curve (area = %0.2f)' % roc_auc)
plt.plot([0, 1], [0, 1], color='grey', lw=2, linestyle='--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver Operating Characteristic')
plt.legend(loc="lower right")
plt.show()
```


### Висновок

На основі ROC-кривої можна зробити висновок, що наша модель Random Forest є високоефективною для даного завдання класифікації. Модель має високу здатність розрізняти між класами, що підтверджується високим значенням AUC (0.98) і крутим підйомом ROC-кривої. Це свідчить про те, що модель добре справляється із завданням і може бути використана для подальших прогнозів.
